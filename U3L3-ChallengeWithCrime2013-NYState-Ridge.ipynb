# Vanilla Logistic Regression

```python
# I know it's wiser to cross-val w/ folds, need to convert
```

```python
# Define the training and test sizes.
trainsize = int(pc_df.shape[0] / 2)
df_train = pc_clean_df.iloc[:trainsize, :].copy()
df_test = pc_clean_df.iloc[trainsize:, :].copy()
```

```python
# Declare a logistic regression classifier.
# Parameter regularization coefficient C to be adjusted.
lr = LogisticRegression(C=1e9)
X_train = df_train[['violent_crime','murder','robbery', 'aggravated_assault',
           'property_crime','burglary','larceny_theft','motor_vehicle_theft',
            'arson']].copy()
Y_train = df_train['greater_than_avg_rape'].copy()
```

```python
X_test = df_test[['violent_crime','murder','robbery', 'aggravated_assault',
           'property_crime','burglary','larceny_theft','motor_vehicle_theft',
            'arson']].copy()
Y_test = df_test['greater_than_avg_rape']
```

```python
# Fit the model.
lr.fit(X_train, Y_train)
```

```python
# Display.
print('Coefficients')
print(lr.coef_)
print('Intercept')
print(lr.intercept_)
pred_y_train = lr.predict(X_train)
pred_y_test = lr.predict(X_test)
print(pred_y_test)
```

```python
pred_y_test = lr.predict(X_test)


```

```python
# Probability Prediction
pred_proba_y_train = lr.predict_proba(X_train)
print(pred_proba_y_train)
```

```python
# Y first
print('\n Accuracy by crime stats stuff- not a 50/50 balanced class!')
print('pred_y_train\n',pd.crosstab(Y_train, pred_y_train))

print('pred_y_test\n', pd.crosstab(Y_test, pred_y_test))

print('\n Percentage accuracy')
print(lr.score(X_test, Y_test))


```

# Lasso Logistic Regression: L1

## Proper notation for training and test set

Y_ # designates the prediction , is the same as y_pred
X and lowercase y

```python
lasso_reg = LogisticRegression(C=3, penalty='l1', tol=0.01)

```

```python
lasso_reg.fit(X_train, Y_train)

```

```python
# Coefficients
l1_coefficients = lasso_reg.coef_.ravel()
print(l1_coefficients)
print(lasso_reg.intercept_)
```

```python

print("score with L1 penalty: %.4f" % lasso_reg.score(X_train, Y_train))
```

# Ridge Regression: L2


```python
ridge_reg = LogisticRegression(C=1, penalty='l2', tol=0.01)
ridge_reg.fit(X_train, Y_train)
l2_coefficients = ridge_reg.coef_.ravel()
print(l2_coefficients)
```

```python
print("score with L2 penalty: %.4f" % ridge_reg.score(X_train, Y_train))
```
